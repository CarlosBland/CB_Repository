```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv("antropometria-dataset-2.csv")
df.head()
```

```python
pivot_q1 = pd.pivot_table(df,
                    index='desc_ent',
                    aggfunc='mean',
                    values ='peso',
                    
                         )
pivot_q1
```

```python
pivot_q2 = pd.pivot_table(df,
                    index='desc_ent',
                    aggfunc='median',
                    values = 'peso',
                    columns='sexo',
                         )
pivot_q2
```

```python
df_adult = pd.read_csv('adult.csv')

df_adult.head()
```
